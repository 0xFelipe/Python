```python
#lista de arquivos no diretorio
os.listdir()
#transformando a pasta em uma variavel contendo todos os arquivos
relatorios = os.listdir()
```

```python
# Setando variavel a cada arquivo (tambem pode ser feito diretamente nessa primeira variavel sem a necessidade de criar outra como no exemplo abaixo)
relatorio1 = pdfplumber.open("/content/gdrive/MyDrive/Colab Notebooks/Modulo 2/Aula 3 Arquivos/20200801.pdf").pages[0]
relatorio2 = pdfplumber.open("/content/gdrive/MyDrive/Colab Notebooks/Modulo 2/Aula 3 Arquivos/20200802.pdf").pages[0]
relatorio3 = pdfplumber.open("/content/gdrive/MyDrive/Colab Notebooks/Modulo 2/Aula 3 Arquivos/20200803.pdf").pages[0]
relatorio4 = pdfplumber.open("/content/gdrive/MyDrive/Colab Notebooks/Modulo 2/Aula 3 Arquivos/20200804.pdf").pages[0]
relatorio5 = pdfplumber.open("/content/gdrive/MyDrive/Colab Notebooks/Modulo 2/Aula 3 Arquivos/20200805.pdf").pages[0]
```

```python
# Setando Variavel da pagina selecionada nesse caso pagina 1
pdf1 = relatorio1.pages[0]
pdf2 = relatorio2.pages[0]
pdf3 = relatorio3.pages[0]
pdf4 = relatorio4.pages[0]
pdf5 = relatorio5.pages[0]
```

```python
# Setando variavel para extract_text() ou seja extrair o texto do arquivo
pdf1 = relatorio1.extract_text()
pdf2 = relatorio2.extract_text()
pdf3 = relatorio3.extract_text()
pdf4 = relatorio4.extract_text()
pdf5 = relatorio5.extract_text()
```

```python
# dividindo o texto, a função split() separa do texto o argumento contido dentro dos parenteses
pdf1.split("\n")
```

```python
# Apartir da chamada [numero] posso escolher qual linha especifica quero separar do texto nesse caso trabalharemos com a linha 6 no qual seria o indce 7 pois contamos apartir do 0
pdf1.split("\n")[6]
```

```python
# Usando a função replace pegamos o que esta dentro da primeira aspas e substituimos pelo conteudo da 2º aspas 
pdf1.split("\n")[6].replace("\t", "")
```

```python
# usando o split() podemos separar o que vem antes com o que vem depois como queremos somente a parte numerica usaremos com R$ para separar o numero do texto
pdf1.split("\n")[6].replace("\t", "").split("R$")
```

```python
# como separamos os valores com split() transformamos em indices entao podemos chamar o indice 1 que no caso seria a 2º palavra apartir do R$
pdf1.split("\n")[6].replace("\t", "").split("R$")[1]
```

```python
# Como pretendendemos trabalhar com valores precisamos converter isso em valores ja que quando foi extraido ele veio em formato texto ou (string) faremos isso colocando tudo entre parentese e chamando a variavel float
float(pdf1.split("\n")[6].replace("\t", "").split("R$")[1])
```

```python
# Apartir da função for variavel in [...]: se cria uma lista
for lista in ["1", "2", "3", "4"]:
# podendo inclusive acrescentar texto ou informaçoes como no exemplo abaixo
  print("Indice", lista)
```

```python
# no exemplo abaixo eu uso a variável soma e atribuo o valor de 0 porem no índice [2] eu atribuo que soma terá o valor do resultado
# + o valor do próximo índice ex 0+1=1... 1+2=3... sucessivamente
soma = 0
for lista in[1, 2, 3, 4]:
  soma = soma + lista
  print(soma)
```

```python
# contador
soma = 0

#criando a função
# Para(arquivos) contidos em "relatório"
for arquivos in relatorios:
  # abrir os arquivos com o pdflumber
  abrir = pdfplumber.open(arquivos)
  #pagina eu seto que quero extrair somente a pagina 1 de todos os pdfs
  pagina = abrir.pages[0]
  # texto mostro que quero que seja extraido apenas o texto do pdf
  texto = pagina.extract_text()
  # valor mostro que após separar "\n" selecionar o índice 6 substituir \t e separar o R$ e selecionar o 2 índice
  valor = texto.split("\n")[6].replace("\t", "").split("R$")[1]
  # transformo o valor em float
  valor = float(valor)
  #somo o resultado de todos e armazeno na variável soma
  soma = soma + valor
  # imprimo somente o valor para controle
  print(arquivos, "--->", valor)
```

```python
# soma total
print("Receita dos Últimos 5 dias: R$", soma)
```
