```python
#Números sem aspas são considerados float
type(295.00)
```

```python
#Qualquer item dentro de aspas é considerado string
str = "R$ 1295,00"
type(str)
```

```python
#Split é usado quando se quer separar grupos pelos espaços no caso RS pertence ao grupo 0 e o valor numerico ao grupo 1
str.split()
```

```python
#Usando a mesma logica de cima selecionamos apenas o grupo 1 o número
numero = str.split()[1]
print(numero)
```

```python
#Usando a variavel replace para trocar a virgula por um ponto
convert = numero.replace(",", ".")
```

```python
#Convertendo a string em um float
float(convert)
```

```python
# Associando Strings as variaveis
nome = "Felipe"
sobrenome = "Mendes"
sufixo = "Souza"
profissao = "Analista de Suporte"

#ver qual é a primeira letra da variavel "nome"
nome[0]
```

```python
#ver qual a ultima letra da variavel nome
nome [-1]
```

```python
# verificar o tamanho do nome
len(nome)
```

```python
virgulas = ("Felipe,Souza,Mendes")
print(virgulas)
```

```python
#replace pega o item nomeado no primeiro indice e subistitui pelo segundo indice 
espaço = virgulas.replace(",", " ")
print(espaço)
```

```python
#Split separa apartir do item citado
separar = virgulas.split(",")
print(separar)
```

```python
#remove espaços em branco ou caracteres indesejados 
Nome = ("       Felipe Souza Mendes                     ").strip()
Nome.strip()
print(Nome)

preço = ("R$ 7,80")
preço = preço.strip("R$")
print(preço)
```

```python
#slicing nos permite "cortar" as strings [0]
nome[0:6]
```

```python
print('Nome Completo: ', nome)
print('Primeira parte: ', nome[:3])
print('Segunda parte: ', nome[3:])
print('Nome feito adicionando Slices: ', nome[:3] + nome[3:])
```

```python
#Imprimir nome completo e profissão
# A virgula conta como espaço enquanto o + apenas soma
print(nome, sufixo, sobrenome)
print("Atua como:" + profissao + ".")
```
