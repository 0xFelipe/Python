```python
# Criando o arquivo usando um Variavel
txt = "Primeiro_Arquivo.txt"
# Abrindo o arquivo com uma das propriedades, W(Escrever), R(Ler), A(Adicionar).
escrever = open(txt, "w")
# Chamando o arquivo pela variavel "escrever" usando .write para esrever dentro
escrever.write("Teste de texto. \nTeste de Paragrafo\nFelipe")
# \n pode ser usado para colocar um paragrafo ou linha
# para fechar o arquivo vamos usar o .close()
escrever.close()
```

```python
# Abrindo o arquivo para leitura 
ler = open(txt, "r")
# Lendo o arquivo se usa .read()
texto = ler.read()
# print para imprimir o texto escrito
print(texto)
# simplificando pode se colocar
print(ler.read())
# fechando
ler.close()
```

```python
# Simplificando a operaçao acima
with open("Segundo_Arquivo.txt", "w") as txt:
  txt.write("Verção Simplificada\nFelipe")
```

```python
# Adicionando texto ao arquivo sem apagar os dados anteriores
with open("Segundo_Arquivo.txt", "a") as txt:
  txt.write("\nAgora o barato ficou louco")
```

```python
# Lendo o arquivo da forma simplificada
with open("Segundo_Arquivo.txt", "r") as txt:
  print(txt.read())
```
